# Two Sample t test

```python
from numpy.random import seed
from numpy.random import randn
from scipy.stats import ttest_ind
```

```python
seed(1)
# generate two independent samples
data1 = 5 * randn(100) + 50
data2 = 5 * randn(100) + 50
```

```python
from scipy import stats
# compare samples
stat, p = ttest_ind(data1, data2)
print('t=%.3f, p=%.3f' % (stat, p))
```

# Paired Sample t test

```python
seed(1)
# generate two independent samples
data1 = 5 * randn(100) + 50
data2 = 5 * randn(100) + 50
```

```python
ttest,pval = stats.ttest_rel(data1, data2)
print(pval)
```

```python
if pval < 0.05:    # alpha value is 0.05 or 5%
   print(" we are rejecting null hypothesis")
else:
  print("we are accepting null hypothesis")
```

# One Sample Z test

```python
from statsmodels.stats import weightstats as stests
```

```python
ztest ,pval = stests.ztest(data1, x2=None, value=50)
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

# Two sample Z test

```python
ztest ,pval = stests.ztest(data1, x2 = data2,value=0,alternative='two-sided')
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

# Anova test

```python
seed(2)
# generate two independent samples
data1 = 5 * randn(100) + 58
data2 = 5 * randn(100) + 50
data3 = 5 * randn(100) + 51
```

```python
F, p = stats.f_oneway(data1,data2,data3)
print("p-value for significance is: ", p)
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
